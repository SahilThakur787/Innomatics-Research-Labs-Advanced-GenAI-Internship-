```python
import pandas as pd

rating_bins = [3.0, 3.5, 4.0, 4.5, 5.0]
rating_labels = [
    '3.0  3.5',
    '3.6  4.0',
    '4.1  4.5',
    '4.6  5.0'
]

final_df['rating_range'] = pd.cut(
    final_df['rating'],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)
```

```python
rating_revenue = (
    final_df
    .groupby('rating_range')['total_amount']
    .sum()
    .sort_values(ascending=False)
)

rating_revenue
```
